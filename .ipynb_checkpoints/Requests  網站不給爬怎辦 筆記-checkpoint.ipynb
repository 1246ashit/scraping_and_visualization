```python
#有些網站不歡迎蟲子
#一般只允許搜尋引擎的蟲子爬
import requests 
url="https://www.kingstone.com.tw/basic/2014150472172?zone=book&lid=home_act_prod2"
r=requests.get(url)
print(r.status_code)#400拒絕存取
```

```python
#利用header
#我們需要偽裝自己
#User-Agent 類似仲介，會相伺服器介紹自己是哪個瀏覽器
import requests 
url="https://www.kingstone.com.tw/basic/2014150472172?zone=book&lid=home_act_prod2"

headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) "}#key =User-Agent #value=你想遞的名信片

r=requests.get(url,headers=headers)
print(r.status_code)
print(r.text)
```

```python
import requests 
from fake_useragent import UserAgent

url="https://www.kingstone.com.tw/basic/2014150472172?zone=book&lid=home_act_prod2"
ua=UserAgent()
header={"User-Agent":ua.random}
r=requests.post(url,headers=header)
print(r.status_code)
print(r.headers)
```

```python
#useragent 可能有限制
#在網頁根目錄的網址再加上 /robots.txt
#找到User-Agent:
#依照他提出的User-Agent來提供
import requests 
target="switch"
#https://shopee.tw/search?keyword=switch
url="https://shopee.tw/search"
param={
    "keyword":target
}
us={"User-Agent":"Googlebot",}
r=requests.get(url,headers=us,params=param)
print(r.url)
print(r.text)
```

```python
#有些網頁會看你來的路徑
#所以要修改Referer裡的網址
import requests
url="http://www.webscrapingfordatascience.com/referercheck/secret.php"
myheaders={"Referer": "http://www.webscrapingfordatascience.com/referercheck/"}
r=requests.get(url,headers=myheaders)
print(r.text)
```

```python
#如果這網站會被重新導向，並且你的爬蟲目標是第一頁的話
#http://www.webscrapingfordatascience.com/redirect/
import requests
url="http://www.webscrapingfordatascience.com/redirect/"
r=requests.get(url,allow_redirects=False)
print(r.text)
print(r.headers)
#requsts 上網的模式類似於其他browser
#可以要求Requests不要轉，利用allow_redirects=False
```

```python
#有些網站是在還沒進入之前就跟你索取帳密
#http://www.webscrapingfordatascience.com/authentication/
#利用auth 第一欄是帳號，第二欄是密碼
url="http://www.webscrapingfordatascience.com/authentication"
r=requests.get(url,auth=("gg","ininder"))
print(r.text)
```

```python
#有些網站會使用cookies
#cookies 簡單而言是用來記錄你在該網頁的行為
import requests
url="http://www.webscrapingfordatascience.com/cookielogin/secret.php"
r=requests.get(url)
print(r.text)
```

```python
#試試加掛cookies
import requests
url="http://www.webscrapingfordatascience.com/cookielogin/secret.php"

mycookies={"PHPSESSID":"ip1asdc5ecltbv552ai3rirlss"}
r=requests.get(url,cookies=mycookies)

print(r.text)
```

```python
#自動化
#要記得先刪掉該網站的cookies
#cookies會跳號
import requests
url="http://www.webscrapingfordatascience.com/cookielogin/"
r=requests.post(url,data={
    "username": "gg",
    "password": "inin"})

#response裡會回傳cookies
print(r.cookies)

r=requests.get(url+"secret.php",cookies=r.cookies)
print(r.url)
print(r.text)
```

```python
#解決直接跳轉的問題
import requests
url="http://www.webscrapingfordatascience.com/redirlogin/"
r=requests.post(url,data={
    "username": "gg",
    "password": "inin"},allow_redirects=False)# allow_redirects=False 是關鍵

#response裡會回傳cookies
print(r.cookies)

r=requests.get(url+"secret.php",cookies=r.cookies)
print(r.url)
print(r.text)
```
