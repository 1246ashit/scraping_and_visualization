```python
#案例 擷取wiki
#requests 使用步驟
#1.載入工具 requests

#2.用requests.get()開啟目標網站

#3.第2步完成後會拿到requests.Response
```

```python
import requests
link="https://www.youtube.com/watch?v=dQw4w9WgXcQ&ab_channel=RickAstley"
aboutKiwi=requests.get(link)
print(type(aboutKiwi))
print("--"*50)

#查看模組
print(dir(aboutKiwi))
print("--"*50)

#查看response裡有啥
text=aboutKiwi.text
print(text)
```

```python
#呼叫該網頁的網址
print(aboutKiwi.url)
```

```python
print(aboutKiwi.status_code)
#200代表連線成功
```

```python
import requests
url="https://www.youtube.com/watch?v=dQw4w9WgXcQ&ab_channel=RickAstley"
html=requests.get(url)

#requests.code.ok==200 
if(html.status_code==requests.codes.ok):
    print("rick roll ")
else:
    print("nice try")
```

```python
#呼叫 headers
html.headers
```

```python
#呼叫 key value par
html.headers["Content-type"]
#印出該網頁的資料格式
```

```python
#呼叫這份回應的請求
html.request.headers
```

```python
#呼叫該網頁編碼模式
html.encoding
```
