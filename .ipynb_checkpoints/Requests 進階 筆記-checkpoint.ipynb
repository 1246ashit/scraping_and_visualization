```python
#get 類似名信片
#大家都看的到，因此安全性不高
#長度有限制，無法送太多資料

#post 類似寄包裹
#重要資料都在封包裡，隱密性夠
#因此可以做很多較複雜的工作
```

```python
#post 用法
import requests
url="http://www.webscrapingfordatascience.com/postform2/"
formdata={
    "name":"benny",
    "gender":"M",
    "pizza":"like",
    "haircolor":"black",
    "comments":"book"
}
r=requests.post(url,data=formdata)
print(r.text)
```
