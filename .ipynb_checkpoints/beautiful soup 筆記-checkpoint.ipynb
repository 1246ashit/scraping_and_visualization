```python
#找到維基百科的參考
import requests
from bs4 import BeautifulSoup
url="https://zh.wikipedia.org/wiki/%E4%B8%B9%C2%B7%E9%9B%B7%E8%AF%BA%E5%85%B9"
r=requests.get(url)
soup=BeautifulSoup(r.text,"lxml")
reference1=soup.find(id="cite_note-1")

reference1
```

```python
#會拔掉所有element，只留下內容
reference1.text
```

```python
#string 跟text不一樣
#如果文本數大餘2，就會回傳null
reference1.string
```

```python
#getText() 跟text一樣
reference1.getText()
```

```python
#子節點
#一個tag可能包含多個string或其他tag，這些都是這個tag的子節點
import requests
from bs4 import BeautifulSoup
url="https://www.wnacg.org/photos-slide-aid-124245.html"
r=requests.get(url)
soup=BeautifulSoup(r.text,"lxml")
r.url
```

```python
#contents和children

#tag的直接子節點

soup.ul.contents
```

```python
#需利用迴圈，才能展示內容
soup.ul.children
```

```python

for i in soup.ul.children:
    print(i)
    print("*"*20)
```

```python
# .descendants 叫出所有後代

for i in soup.ul.descendants:
    print(i)
    print("="*40)
```
