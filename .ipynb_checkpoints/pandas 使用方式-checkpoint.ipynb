```python
from bs4 import BeautifulSoup
import requests

url="https://www.imdb.com/chart/top/?ref_=nv_mv_250"
r=requests.get(url)
soup=BeautifulSoup(r.text,"lxml")

rank=soup.find("table",class_="chart full-width")
for i in rank.find("tbody").findAll("tr"):
    name=i.find("td",class_="titleColumn").find("a")
    print(name.text)
    
    year=i.find("span",class_="secondaryInfo")
    print(year.text[1:5])
    
    rating=i.find("strong",title=True)
    rating=rating.get("title").split(" ")
    print("評分:",rating[0])
    print("評分人數:",rating[3])
    print("="*20)

  
```
