功能跟findAll類似，選取規則依照CSS規則，又叫做CSS選擇器


```python
#透過標籤名查詢
import requests
from bs4 import BeautifulSoup
url="https://zh.wikipedia.org/wiki/Vivy_-Fluorite_Eye%27s_Song-"
html=requests.get(url)
soup=BeautifulSoup(html.text,"lxml")

soup.select("p")
```

```python
soup.select("a")
```

```python
# #value 查詢ID
soup.select("#mw-page-base")
```

```python
# .value 查詢類別
soup.select(".noprint")
```

```python
#透過屬性查詢
soup.select("[href='/wiki/%E7%A8%AE%E5%B4%8E%E6%95%A6%E7%BE%8E']")
```
