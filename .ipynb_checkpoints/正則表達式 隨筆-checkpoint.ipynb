1.載入re
2.創建模具
3.過濾文件
4.用group來印出結果

```python
#正則表達式

import re  #載入模組 re

test="my phonenumber is 0909777777 and 0909888888"

foundNum=re.compile(r"\d{10}") #compile用來建模具， r直接過濾單字，r要記得放在分號的前面

foundNum.findall(test) #findall 用來過濾出答案
```

```python
#用來看re裡的工具有哪些
dir(re)
```

```python
#\d 用來過濾數字

foundNum=re.compile(r"\d")
foundNum.findall(test)
```

```python
# split 切割單字

findWord=re.split("r\s",test)
findWord
```

```python

text="my phonenumber is 090-977-7777"
findNum=re.compile(r"\d{3}-\d{3}-\d{4}") # "\d{3}"更快速告知要幾個數字  #value可儲存模具

mo=findNum.search(text) #valueName.search()啟用模具，括號裡可放入要過濾的東西 #模具搜尋完的結果可放入value中


print("比較好讀的結果:",mo.group())#使用valueName.group() 來叫出比較好讀的結果

print("不太好讀的結果:",mo)
```

```python
#可以一句寫好
text="我的電話是0909777777"

answer=re.search(r"\d{10}",text)  #直接呼叫re.search() 模具在括號裡第一個位置建立，要過濾的內容則在第二個位置

print(answer.group(),"是我的電話")
#以上寫法只能找到第一個搜到的結果
```

```python
#這寫法可找到所有符合結果
text="我有兩支電話 一隻是 0909777777 第二隻是 0909888888"
findNum=re.compile(r"\d{10}")
findNum.findall(text)
```

```python
#利用()來群組
text="我的號碼是 0909777777"
findNum=re.search(r"(\d{3})(\d{7})",text)#格式為 "(\d{})"

print("印出前三個數字:",findNum.group(1))
print("印出後七個數字:",findNum.group(2))
print("印出全部的數字:",findNum.group())#預設為全印
```

```python
text="我的電話號碼為 02-1234-8888"
ans=re.search(r"(\d{2})-(\d{4}-\d{4})",text)
print("我的公司電話是"+ans.group(2)+","+"外縣市請加"+ans.group(1))
print()
print("我的公司電話是%s"%ans.group())
```

```python
#要搜尋特殊字元的話
#在該特殊字元前加上\

text="電話號碼是 (090)9777777"
ans=re.search(r"\(\d{3}\)\d{7}",text)
ans.group()

#在模具最前面記得寫上r，會方便很多
```

```python
#利用 | 來挑選其一
#會先找第一個條件，之後再找第二個
import re

mo=re.compile(r"I|have")
text="I have 30cm"
findWord=mo.search(text)
print(findWord.group())

text2="have 30cm"
findWord=mo.search(text2)
print(findWord.group())
```

```python
#如果要一次找到 水之呼吸，火之呼吸，雷之呼吸

mo=re.compile(r"(水之|火之|雷之)呼吸")
text="炭之狼會火之呼吸"
findWord=mo.search(text)
print(findWord.group())
print(findWord.group(1))
```

```python
# ? 的用法: 有或沒有都要
text="我不是女的"
mo1=re.compile(r"我(不)?是女的")
ans=mo1.search(text)
print(ans.group())
print(ans.group(1))
```

```python
# * 的用法: 就算這詞沒出現也要
text="我沒有,抱歉"
mo1=re.compile(r"我沒有(筆)*,抱歉")

ans=mo1.search(text)
print(ans.group())
```

```python
# + 的用法:至少出現一次才要
text1="我沒有筆"
mo1=re.compile(r"我沒有(筆)+")
ans=mo1.search(text1)
print(ans.group())
```

```python
#如果希望pattern可以重複次數
#(pattern){digital}
text="好嘞拉拉拉拉拉"
mo=re.compile(r"好嘞(拉){2,6}")
ans=mo.search(text)
print(ans.group())
```

```python
#python 的 reguler expreesions 有兩種模式
#greed & lazy 用 ? 號來切換

text=" 阿阿阿阿阿阿阿"

mo=re.compile(r"(阿){1,7}")#預設為greed
ans=mo.search(text)
print(ans.group())
print(len(ans.group()))

print() 

mo=re.compile(r"(阿){1,7}?")#切換為lazy
ans=mo.search(text)
print(ans.group())
print(len(ans.group()))
```

```python
#search 只能找到第一個符合的
#findall 能找到全部

t="我有5公分，我妹有30公分"
mo=re.compile(r"\d{1,2}公分")
ans=mo.findall(t)
print(ans)
```

```python
#字元族
#\d 數字
#\s 空白
#\w 數字，文字，_
mo=re.compile(r"\d+\s\w+")
t="12 apple,好吃,2 碗白飯"
ans=mo.findall(t)
print(ans)
```

```python
#自訂自己的模型  []
mo=re.compile(r"[s,e,x]")
t="six elephan there are taking x-ray"
ans=mo.findall(t)
print(ans)
```

```python
#想要做一個區間
#[a-z] [A-Z] [0-9]
#要加入特殊符號可直接加
#也能加上重複次數
mo=re.compile(r"[1-4\a-z]{2}")
t="1dada2fdgg3r\wqw4sdfs"
ans=mo.findall(t)
print(ans)
```

```python
# ^ 符號:放在方括號是尋找這字以外的
mo=re.compile(r'[^hello]')
t="hello你這個臭雞雞"
mo.findall(t)
```

```python
# ^ 放在括弧裡是要求開頭就是要這字串
mo=re.compile(r"^hello")

t="hello你這個臭雞雞"
print(mo.findall(t))

t="你這個臭雞雞hello"
print(mo.findall(t))
```

```python
# $ 放在括弧裡是要求結尾就是要這字串
mo=re.compile(r"hello$")

t="hello你這個臭雞雞"
print(mo.findall(t))

t="你這個臭雞雞hello"
print(mo.findall(t))
```

```python
# $ ^ 放在括弧裡是要求全部符合
mo=re.compile(r"^hello$")

t="hello你這個臭雞雞hello"
print(mo.findall(t))

t="你這個臭雞雞hello"
print(mo.findall(t))

t="hello"
print(mo.findall(t))
```

```python
# . 符號: 外卡符號，放前面
mo=re.compile(r".雞雞")
t="臭雞雞,大香雞雞,小雞雞雞,公雞"
print(mo.findall(t))
```

```python
# .+ 符號: 一個字以上
mo=re.compile(r".+雞雞")
t="臭雞雞,大臭雞雞,小小雞雞,公雞"
print(mo.findall(t))
```

```python
# .* 萬物符號
mo=re.compile(r"姓:(.*)名:(.*)")
t="姓:捷克 名:曼杰森"
ans=mo.findall(t)
print(ans)
```
