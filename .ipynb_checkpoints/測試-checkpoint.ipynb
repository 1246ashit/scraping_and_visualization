```python
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys

url = 'https://www.momoshop.com.tw/category/MgrpCategory.jsp?m_code=3900100014'
driver = webdriver.Chrome()
driver.get(url)

target=driver.find_element_by_link_text("朵舒")
target.click()

paper=driver.find_elements_by_xpath('//li[@class="eachGood"]')
for detail in paper:
    detail=detail.find_element_by_xpath('//*[@class="prdPrice"]')
    print(detail.text)
    print("="*20)
                     
```
