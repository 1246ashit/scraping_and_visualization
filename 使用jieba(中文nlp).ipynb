```python
#Jieba 透過 TF-IDF 找出句子關鍵字
import jieba.analyse
jieba.load_userdict('./trandition_word_dict.txt')

news="蔡總統是在早上7點30分左右到現場，\
    依照相關流程，報到、問診、接種疫苗，\
    接種完畢後將會交由醫療團隊後續掌握身體狀況。\
    總統施打完疫苗後，也將返回官邸，\
    與東京奧運選手進行第三梯次，\
    也是近期最後一場次的視訊會議。"

tags = jieba.analyse.extract_tags(news, topK=20, withWeight=True)

'''
sentence 為句子
topK 代表返回 TF-IDF 權重最大的關鍵字，默認為 20
withWeight 代表是否返回關鍵字權重值，默認為 False
allowPOS 代表指定詞性，默認為空，也就是不篩選
'''

tags
```

# 去除停用詞

```python
#載入停用字
stopword=[]
with open('./my_stopwords.txt',"r",encoding="utf8") as file:
    for i in file.read():
        if(i!="\n"):
            stopword.append(i)
        else:
            continue
def check_stopword(word):
    if(word not in stopword):
        return word 
print("stopword is get ready for work")
```

```python
##去除停用字
import jieba.posseg as pseg
import jieba

jieba.load_userdict('./trandition_word_dict.txt')
news="蔡總統是在早上7點30分左右到現場，依照相關流程，報到、問診、接種疫苗，接種完畢後將會交由醫療團隊後續掌握身體狀況。總統施打完疫苗後，也將返回官邸，與東京奧運選手進行第三梯次，也是近期最後一場次的視訊會議。"

fixed_str1=jieba.lcut(news)

no_stop_word=""
for i in fixed_str1:
    if(check_stopword(i)==None):
        continue
    no_stop_word+=(check_stopword(i))
print(no_stop_word)
fixed_str1
```

```python
#去掉停用詞後在偵查詞頻
import jieba.analyse
jieba.load_userdict('./trandition_word_dict.txt')


tags = jieba.analyse.extract_tags(no_stop_word, topK=20, withWeight=True)
tags
```
