```python
#爬momo
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import re

url = 'https://www.momoshop.com.tw/category/MgrpCategory.jsp?m_code=3900100014'
driver = webdriver.Chrome()
driver.get(url)
ac=ActionChains(driver)

name=[]
price=[]
brand=[]


#擷取各品牌的商品
#共有22個廠牌
brandlist=["朵舒","舒潔","春風","TEMPO","柔情","得意",\
           "優活","優朵","綠荷","統潔","五月花","百吉牌",\
           "倍潔雅","唯潔雅","蒲公英","新新新","簡約組合",\
           "情人 / 雪柔","可麗舒/可立雅","極度純柔","日本大王"]
    
    
#爬朵舒的產品
for i in brandlist:
    try:
        driver.find_element_by_link_text(i).click()
        soup=BeautifulSoup(driver.page_source,"lxml")
        paper=soup.find("div",{'class':'prdListArea bt770class'})
      #紀錄資料
        for detail in soup.find_all("li",{"class":"eachGood"}):
            name.append(detail.find("h3",{"class":"prdName"}).text)#名子
            
            #過濾出價錢數字
            price_str=detail.find("span",{"class":"prdPrice"}).text
            mo=re.search(r'\d+,\d+|\d+',price_str)
            mo=mo.group()
            mo=mo.replace(",","")
            price.append(eval(mo))#價錢
            
            brand.append(i)#品牌
            print("-"*20)
        print(i,"完成")
    except:
        driver.find_element_by_link_text("更多").click()
        driver.find_element_by_link_text(i).click()
        soup=BeautifulSoup(driver.page_source,"lxml")
        paper=soup.find("div",{'class':'prdListArea bt770class'})
      #紀錄資料
        for detail in soup.find_all("li",{"class":"eachGood"}):
            name.append(detail.find("h3",{"class":"prdName"}).text)#名子
            
            #過濾出價錢數字
            price_str=detail.find("span",{"class":"prdPrice"}).text
            mo=re.search(r'\d+,\d+|\d+',price_str)
            mo=mo.group()
            mo=mo.replace(",","")
            price.append(eval(mo))#價錢
            
            brand.append(i)#品牌
            print("-"*20)
        print(i,"完成")
driver.quit()        
print("爬蟲結束")
```

```python
#資料整理成excel 
import pandas as pd

toilet_paper= pd.DataFrame({'品牌':brand,
                    '品項名子':name,
                    '價格':price
                        
                      })

toilet_paper.to_csv("toilet_paper.csv")
toilet_paper.to_excel("toilet_paper.xlsx")

toilet_paper
```
