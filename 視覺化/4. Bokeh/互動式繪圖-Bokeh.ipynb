## <span style="color:red ">Bokeh的建造概念
------
    1. Application-這是你最後呈現在伺服器上面的整個結果
    2. Plot-圖的主體，就是你在畫的這個圖!!
    3. Glyphs-就是你畫在圖上的圓圈圈、線啊!!
    4. Widgets-就是將來我們要建造的那些互動部件

## 畫圖的步驟
    1. 準備資料-Prepare some data
    2. 告訴Bokeh要把圖輸出到哪裡?
       - output_file(), 就是輸出一個html檔案，可以自己指定檔案名 
       - output_notebook()，就是要在Jupyter notebooks裡面顯示

    3. Call figure()
       -把畫紙攤出來，可以增加title, tools, and axes labels.各式各樣的相關配件  

    4. 開始畫圖-Add renderers
       -比如p.line(x,y)，就會畫一個線圖

    5. 請 Bokeh  show() or save() 把結果show出來

### <span style="color:red ">bar plots-你希望x軸的排列可以依照一個遞增或是遞減的趨勢...

````python
from bokeh.plotting import figure,output_notebook, show

output_notebook()

#先做一組數據出來
fruits =['蘋果', '桃子', '鳳梨', '香蕉', '藍莓', '火龍果']
counts = [5, 3, 4, 2, 4, 6]

#  其實很簡單，就用python的sorted功能，把x軸排列一次，再告訴bokeh
#  reverse=True
sorted_fruits = sorted(fruits, key=lambda x: counts[fruits.index(x)])
p = figure(x_range=sorted_fruits, plot_height=250, title="Fruit Counts")

p.vbar(x=fruits, top=counts, width=0.5)

p.xgrid.grid_line_color = None
p.y_range.start = 0

show(p)
````

### <span style="color:red ">bar plots-怎麼調整顏色?

````python
from bokeh.plotting import figure,output_notebook, show
#導進顏色模組
from bokeh.palettes import Category10

output_notebook()
#先做一組數據出來
x = [8,9,10]
y = [1,2,3]

plot = figure(plot_width=200, plot_height=200)

#vbar是vertical bar的意思
#如果你可能對英文有點概念，你會想到hbar
#把色盤放進來，顏色數要配合資料點數，Category10[3]就是Category10裡面的三色組
plot.vbar(x,top = y, color = Category10[3] , width= 0.5)

#如果要把Y baseline拉平
#plot.y_range.start = 0

#如果你不想要看到x軸的格線
#plot.xgrid.grid_line_color = None

show(plot)
````

### <span style="color:red ">如果你覺得要給顏色個數很煩...-CategoricalColorMapper- factor_cmap()

````python
import pandas as pd
from bokeh.io import output_notebook, show
from bokeh.palettes import Category10
from bokeh.plotting import figure
#導入以下工具factor_cmap
from bokeh.transform import factor_cmap

output_notebook()

fruits = ['蘋果', '桃子', '鳳梨', '香蕉', '藍莓', '火龍果']
counts = [5, 3, 4, 2, 4, 6]

#factor_cmap()這個方法要求要用Dataframe，用等一下要講的ColumnDataSource(bokeh自己的資料格式)，也可以
source = pd.DataFrame({'fruits':fruits,'counts':counts})

p = figure(x_range=fruits, plot_height=250, title="Fruit Counts")

#直接把factor_cmap帶進來，需要設定三個變數
#source裡面的欄位名稱、色盤名稱、決定配對的因素(挑出unique值配發顏色))
p.vbar(x='fruits', top='counts', width=0.9, source=source, legend_field="fruits",
       line_color='white', color=factor_cmap('fruits', palette=Category10[10], factors=sorted(source.fruits.unique())))

p.xgrid.grid_line_color = None
p.y_range.start = 0
p.y_range.end = 9
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

show(p)
````

## <span style="color:red ">跟pandas GroupBy是好朋友

````python
from bokeh.plotting import figure,output_notebook, show

#要顏色，把下面的工具套進來
from bokeh.palettes import Spectral5
from bokeh.transform import factor_cmap
#導入資料
from bokeh.sampledata.autompg import autompg_clean as df


output_notebook()

#等一下factor_cmap會用到cyl，要求要是str，所以先轉一下
df.cyl=df.cyl.astype(str)

#這是一個跟汽車相關的資料集
#'cyl'汽缸數
#'MPG' mile per gallon
group = df.groupby('cyl')


#用的是'cyl'汽缸數這個欄位，factors由這個欄位裡面的unique值決定
cyl_cmap = factor_cmap('cyl', palette=Spectral5, factors=sorted(df.cyl.unique()))

p = figure(plot_height=350, x_range=group, title="MPG by # Cylinders")

#line_color是每一個長條的邊框
#可以自動帶進分組資料
#用分組的mean當成top顯示數據
p.vbar(x='cyl', top='mpg_mean', width=.5, source=group,
       line_color='red', color=cyl_cmap)

p.y_range.start = 0
p.xgrid.grid_line_color = None
p.xaxis.axis_label = "汽缸耗油分析"
p.xaxis.major_label_orientation = 1.2
p.outline_line_color = None

show(p)
````

## <span style="color:red ">GroupBy multi-level也沒有問題-多變數分組

````python
from bokeh.plotting import figure,output_notebook, show
from bokeh.palettes import Spectral5
from bokeh.sampledata.autompg import autompg_clean as df
from bokeh.transform import factor_cmap

output_notebook()

df.cyl=df.cyl.astype(str)
group = df.groupby(by=['cyl', 'mfr'])
#我們現在靠兩個軸位做分類，[cyl,mfr]
#所以我們要靠start和end告訴色盤，要依據甚麼顯示
#default start=0，所以我們配上end=1(包頭不包尾)，就會用cyl配色
index_cmap = factor_cmap('cyl_mfr', palette=Spectral5, factors=sorted(df.cyl.unique()),end=1)

p = figure(plot_width=800, plot_height=300, title="MPG 平均數 by 汽缸數和製造商進行分類",
           x_range=group, toolbar_location=None, tooltips=[("MPG", "@mpg_mean"), ("Cyl, Mfr", "@cyl_mfr")])

p.vbar(x='cyl_mfr', top='mpg_mean', width=1, source=group,
       line_color="white", fill_color=index_cmap, )


p.y_range.start = 0
p.x_range.range_padding = 0.05
p.xgrid.grid_line_color = None
p.xaxis.axis_label = "依汽缸數和製造商進行分類"
p.xaxis.major_label_orientation = 1.2
p.outline_line_color = None

show(p)
````

## <span style="color:red ">吃 ColumnDataSource 來畫圖  
ColumnDataSource是Bokeh的特殊資料格式
------
### 但是要用很簡單

````python
from bokeh.plotting import figure,output_notebook, show
from bokeh.plotting import ColumnDataSource
import pandas as pd

output_notebook()

plot = figure(plot_width=400, plot_height=400)

#把檔案讀進來
df = pd.read_csv('all_stocks_5yr.csv')

#時間資料轉換一下
df['date'] = pd.to_datetime(df['date'])

#只把蘋果的資料篩出來
df_apple = df[df['Name'] == 'AAL']

#創造ColumnDataSource object
#簡單的說，就是把data分配給一個一個新的key，也就是創造一個新的dictionary，那你就完成了
#Bokeh的某些功能，你就是需要ColumnDataSource的幫忙 
data = ColumnDataSource(data = {
    'x' : df_apple['high'],
    'y' : df_apple['low'],
    'x1': df_apple['open'],
    'y1': df_apple['close'],
    'x2': df_apple['date'],
    'y2': df_apple['volume'],   
})


#Create the first scatter plot

plot1 = figure()

plot1.cross(x = 'x', y = 'y', source = data, color = 'red', size = 10, alpha = 0.8)

show(plot1)
````

````python
import numpy as np
from bokeh.layouts import gridplot
from bokeh.plotting import figure, output_notebook, show

# 樣本資料準備
N = 100
x = np.linspace(0, 4*np.pi, N)
y0 = np.sin(x)
y1 = np.cos(x)
y2 = np.sin(x) + np.cos(x)


output_notebook()

#做一個新的plot S1
s1 = figure(width=250, plot_height=250, title='S1')
s1.circle(x, y0, size=10, color="navy", alpha=0.5)

#再做一個新的plot S2， S2和S1會共用X Y range
s2 = figure(width=250, height=250, x_range=s1.x_range, y_range=s1.y_range, title='S2')
s2.triangle(x, y1, size=10, color="firebrick", alpha=0.5)

#再做一個新的plot S3， S3和S1 只共用X range
s3 = figure(width=250, height=250, x_range=s1.x_range, title='S3')
s3.square(x, y2, size=10, color="olive", alpha=0.5)

# 把所有圖放進 gridplot
p = gridplot([[s1, s2, s3]], toolbar_location=None)

# show the results
show(p)
````

# <span style="color:red ">HoverTool
-----
## [tool設定](https://docs.bokeh.org/en/latest/docs/user_guide/tools.html)
   

```
'$'號是Bokeh自帶欄位的資訊  
      $index 
      $name  
      $x  
      $y   
      $color 
'@'號是對應data裡面的欄位
```

````python
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import ColumnDataSource

output_notebook()

source = ColumnDataSource(data=dict(
    x=[1, 2, 3, 4, 5],
    y=[2, 5, 8, 2, 7],
    desc=['A', 'b', 'C', 'd', 'E'],
))

TOOLTIPS = [
    ("index", "$index"),
    ("(x,y)", "($x, $y)"), #不要誤會，這裡的xy不是上面的xy，是圖上座標、用圖的xy標出來
    ("desc", "@desc"),
]

p = figure(plot_width=400, plot_height=400, tooltips=TOOLTIPS,
           title="Mouse over the dots")

p.circle('x', 'y', size=20, source=source)

show(p)
````

## <span style="color:red ">js_link-把Widget和glyph連結在一起

````python
from bokeh.layouts import column
from bokeh.models import Slider
from bokeh.plotting import figure, show,output_file

output_file("line.html")

plot = figure(plot_width=200, plot_height=200)
r = plot.circle([1,2,3,4,5,], [3,2,5,6,4], radius=0.2, alpha=0.5)

slider = Slider(start=0.1, end=2, step=0.01, value=0.2)

#靠js_link，把滑桿的value和圖的property連在一起
slider.js_link('value', r.glyph , 'radius')

show(column(plot, slider))
````

## <span style="color:red ">有時候，你需要的不是部件連結可以解決的-那你就需要加一下Callback
------
    1.Callback你就把他想成當你按鈕按下去，會發出一個命令叫系統去算一個東西，或是換一個畫面。
    2.Bokeh主要透過兩個東西呼叫Callback，.on_change or .on_click

````python
from bokeh.plotting import figure,output_notebook, show
from bokeh.models import Button
import random


output_notebook()

#基本上下面這個寫法是為了要在純python的環境下跑，所以要包成一個function

def bkapp1(doc):
    
    #設計我們的callback function
    # (attr, old, new)是規定的參數
    def callback():
        target=random.randint(1, 11)
        print('恭喜你!抽到{}號'.format(target))
        
    #設計我們的widget
    button = Button(label="按我，抽個號碼!", button_type="success")
    
    #關鍵就是這句，on_change
    button.on_click(callback)

    doc.add_root(button)
        

show(bkapp1)
````

````python
from bokeh.plotting import figure,output_notebook, show
from bokeh.models import TextInput

output_notebook()

#基本上下面這個寫法是為了要在純python的環境下跑，所以要包成一個function

def bkapp1(doc):
    
    #設計我們的callback function
    # (attr, old, new)是規定的參數
    def my_text_input_handler(attr, old, new):
        print('上一個是{}'.format(old))
        print('原來你叫{}'.format(new))
        
    #設計我們的widget
    text_input = TextInput(value="Jeff", title="請輸入你的姓名:")
    
    #關鍵就是這句，on_change
    text_input.on_change("value", my_text_input_handler)

    doc.add_root(text_input)
        

show(bkapp1)
````
