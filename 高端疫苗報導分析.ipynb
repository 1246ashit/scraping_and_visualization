```python
#爬新聞 google 到的第一頁
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys

url = 'https://www.google.com'
driver = webdriver.Chrome()
driver.get(url)


target=driver.find_element_by_xpath('/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input')
target.send_keys("半導體")
target.submit()
driver.find_element_by_link_text("新聞").click()
target=driver.find_elements_by_xpath('//a[@style="text-decoration:none;display:block"]')

news_href=[]
for link in target:
    news_href.append(link.get_attribute("href"))
print(news_href)#要處理的網址    

driver.quit()
```

```python
from selenium import webdriver
driver = webdriver.Chrome()
page=0
artical=[]
for i in news_href:
    page+=1
    print("處理第",page,"頁中")
    driver.get(i)
    news=driver.find_elements_by_xpath("*//p")
    for i in news:
        artical.append(i.text)
        print(i.text)
    print("第",page,"頁完成")
print("文章擷取完成")
#driver.quit()
```

```python
artical="".join(artical)
artical
```

```python
import jieba.analyse
# trandition_word_dict.txt 繁中字典
jieba.load_userdict('./trandition_word_dict.txt')#自訂義字典

keywords = jieba.lcut(artical)

#去除停用字
stopword=[]
with open('./my_stopwords.txt',"r",encoding="utf8") as file:
    for i in file.read():
        if(i!="\n"):
            stopword.append(i)
        else:
            continue
def check_stopword(word):
    if(word not in stopword):
        return word 
print("stopword is get ready for work")

no_stop_word=[]
for i in keywords:
    if(check_stopword(i)==None):
        continue
    no_stop_word.append(check_stopword(i))
#print(no_stop_word)

dict1={}
for i in no_stop_word:
    try:
        dict1[i]!=0
        dict1[i]+=1
    except:
        dict1.setdefault(i,1)
print(dict1)
```

```python
from wordcloud import WordCloud
#把這朵雲的基礎做出來
#下面的width=1600, height=800是設定解析度

font = 'SourceHanSansTW-Regular.otf'#設定中文字體

wcloud =\
WordCloud(background_color='white',\
          
          width=1600, height=800,\
          
          font_path=font).generate_from_frequencies(dict1)
```

```python
import matplotlib.pyplot as plt
plt.figure( figsize=(20,10) )
plt.imshow(wcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
